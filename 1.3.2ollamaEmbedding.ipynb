```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="gemma:2b")#by default its llama2
)
```

```python
text=[
    'Alpha is the first character of the Greek alphabet.',
    'Beta is the second letter of the Greek alphabet.'
]
embeddingtext=embeddings.embed_documents(text)
```

```python
embeddingtext
```

```python
embeddingtext[0]
```

```python
query = ['What is the second letter of Greek alphabet?']
embeddingquery=embeddings.embed_query(query)
embeddingquery
```

```python
len(embeddingquery)
```
