```python
import os
from langchain_community.document_loaders import GithubFileLoader

loader = GithubFileLoader(
    repo="shafqatameen/opencv",
    access_token=os.getenv("GITHUB_PERSONAL_ACCESS_TOKEN"),
    github_api_url="https://api.github.com",
    
    # LOAD MULTIPLE TYPES:
    # This will grab all Markdown, Python, and JavaScript files
    file_filter=lambda file_path: file_path.endswith((".md", ".py", ".js")) 
)

documents = loader.load()

# --- Verification ---
# Let's see what we got
print(f"Loaded {len(documents)} files.")
for doc in documents[:6]:
    # This prints the filename so you can see different types being loaded
    print(f" - Found file: {doc.metadata['source']}")
```
